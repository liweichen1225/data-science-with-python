```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')

%matplotlib inline
```

```python
# Use pandas to read in 'wine_dataset.csv'
# This is a dataset of various wines with a target of categorical variable 'class'
df = pd.read_csv('../data/wine_dataset.csv')
```

```python
# 1. Print out the number of rows and columns in the dataset using shape
df.shape
print('n rows: {}, n columns: {}'.format(*df.shape))
```

```python
# 2. Print out the first 5 rows of df using head
df.head(5)
```

```python
# 3. Print out the summary stats of numeric columns using df.describe()
df.describe()
```

```python
# 4. Using iloc, print out the first 3 rows, first 3 columns
df.iloc[0:1]
df.loc['ash']
```

```python
# 5. Using loc, print out rows with index 4 to 6 inclusive and columns 'ash' and 'total_phenols'
df.loc[4:6].loc[:,['ash', 'total_phenols']]
```

```python
# 6. Print out all rows with hue greater than the median value of hue
# This should result in 89 rows x 14 columns
df[df.hue > df.hue.median()]
#df.loc[df.hue>np.median(df.hue)]
```

```python
# 7.1. Create two axes using plt.subplots with 1 row , 2 columns, figsize=(10,4)
fig,ax =plt.subplots(1,2,figsize=(10,4))

# 7.2 In the first axis (ax[0]), plot the distribution of df.alcohol using sns.distplot with rug=True
sns.distplot(df.alcohol,ax=ax[0],rug = True)

# 7.3 Add the title 'Wine Alcohol Content' to ax[0] using set_title
ax[0].set_title('Wine Alcohol Content')

# 7.4 In the second axis, plot the distribution of df.color_intensity using sns.distplot with rug=True
sns.distplot(df.color_intensity,ax=ax[1],rug = True)

# 7.5 Add the title 'Wine Color Intensity' to ax[1] using set_title
ax[1].set_title('Wine Color Intensity')
```
